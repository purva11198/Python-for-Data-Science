```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('datasets/weather_by_cities.csv')
df
```

### pivot is used for multilvel indexing 

```python
df.pivot(index="day",columns="city")
```

```python
df.pivot(index="day",columns="city",values="temperature")
```

### pivot table is used to summarize and aggregate data inside dataframe

```python
pdf=df.pivot_table(index="city",columns="day") #by default aggregation is mean
pdf
```

```python
df.pivot_table(index="city",columns="day",aggfunc="count") #similarly diff,sum,mean
```

```python
df.pivot_table(index="city",columns="day",margins=True)
```

### To convert back we will use stack method

```python
df = pdf.stack().reset_index()
df
```

```python
pd.crosstab(df.day,df.city[0:11],margins=True,normalize=True)    #margins will add total column
```

### reshape dataframe using melt

```python
df1=pd.melt(df, id_vars=['day'], value_vars=['city'])
df1
```

```python
df1[df1["variable"]=="city"]
```

```python
df1=pd.melt(df,id_vars=["day"],var_name="city",value_name="temperature")
df1
```
